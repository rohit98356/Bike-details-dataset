```python
data=df['km_driven']
outliers = []
def detect_outliers(data):
    threshold = 3
    mean = np.mean(data)
    std = np.std(data)

    for i in data:
        z_score = (i-mean)/std
        if np.abs(z_score)>threshold:
            outliers.append(i)
    return outliers
```

```python
outliers = detect_outliers(data)
```

```python
outliers
```

```python
df[~df['km_driven'].isin(outliers)]
```

```python
#12.Perform a bivariate analysis to visualize the relationship between year and selling_price?
fig = plt.figure(figsize=(15,10))
sns.barplot(x='year',y='selling_price',data=df)
plt.show()
```

```python
#13. What is the average depreciation in selling price based on the bike's age (current year - manufacturing year)?
df['age'] = 2025-df['year']
df.groupby('age')['selling_price'].mean()
```

```python
#14. Which bike names are priced significantly above the average price for their manufacturing year?
average_price = df.groupby('year')['selling_price'].mean()
df['average_price'] = df['year'].map(average_price)
df_above_avg = df[df['selling_price'] > 1.5 * df['average_price']]
print(df_above_avg[['name', 'year', 'selling_price', 'average_price']])
```

```python
#15.Develop a correlation matrix for numeric columns and visualize it using a heatmap.
numeric_columns = df.select_dtypes(include=['int64', 'float64'])
correlation_matrix = numeric_columns.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
